```python
# Computing the Average score 
st["average_score"] = st[['math_score','reading_score', 'writing_score']].mean(axis=1)
st
```

```python
# Assigning grades with letters
def grades(avg):
    if avg >= 75:
        return "A"
    elif avg >= 65:
        return "B"
    elif avg >= 55:
        return "C"
    elif avg >= 45:
        return "D"
    elif avg >= 40:
        return "E"
    else:
        return "F"
st["grade"] = st.average_score.apply(grades)
st
```

```python
# Checking for the top Students with respect to grades
best_students = st.loc[st.grade == 'A']
best_students
```

```python
# Checking for the top Students with respect to maximum average
max_avg = st.average_score.max()
top_students = st[st['average_score'] == max_avg]
top_students
```
